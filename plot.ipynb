```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('saved_results_degree_lr_0.001_ns_del_4_o2_1.csv')

# Plot loss curve
plt.figure(figsize=(10, 5))
plt.plot(data["episode"], data["avg loss"], label='Loss')
plt.xlabel('Episodes')
plt.ylabel('Average Loss')
plt.title('Loss Curve')
plt.legend()
plt.grid(True)
plt.show()

# Plot eigenvalues
plt.figure(figsize=(10, 5))
plt.plot(data["episode"], data["eigenval"], color='r', label='Eigenvalues')
plt.xlabel('Episodes')
plt.ylabel('Eigenvalues')
plt.title('Eigenvalues')
plt.legend()
plt.grid(True)
plt.show()
```
